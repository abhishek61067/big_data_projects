```python
x = np.array([6,8,10,14,18])
y = np.array([7,9,13,17.5,18])

#plotting two graphs simulataneously
m,b = np.polyfit(x,y,deg= 1)
plt.plot(x, m*x+b) #this will show a linear line, coz y= mx+b is a linear equation
plt.plot(x,y,'o') #x,y coordinates
plt.show()
```

```python
variance = np.var(x, ddof = 1)
```

```python
#diameter of pizza is denoted by variable x
x = np.array([6,8,10,14,18])
#diameter of price is denoted by variable y
y = np.array([7, 9, 13, 17.5,18])


#calculating covariance:
covariance = np.cov(x,y)
covariance


#QUESTION?
#WHY IS COVARIANCE PRINTING 4 ELEMENT ENCLOSED IN AN ARRAY? SHOULDNOT THE COVARIANCE SHOW THE DIRECTION OF TWO VARIABLES?
```

```python
covariance[0][1]
```

```python

#calculating beta 
beta = covariance[0][1]/variance
```

**y = beta*X +alpha**
alpha = y-beta * X

```python
x_bar = np.mean(x)
x_bar
```

```python
y_bar = np.mean(y)
y_bar
```

```python
#calculating alpha
alpha = y_bar - beta *x_bar
alpha
```

```python
# so our model will be : y = alpha+ beta *x
y = alpha + beta * x
```

```python
y = alpha +beta * (11)
print(y)
print(alpha+beta*12)
```
