```python
import math
import pandas as pd
```

```python
def gini(samples):
    sum_samples = sum(samples)
    gini_score = 0
    probabilities = []
    for sample in samples:
        probability = sample / sum_samples
        probabilities.append(probability)
    for probability in probabilities:
        gini_score += probability * (1 - probability) 
    return gini_score
```

```python
samples = [2,2] #class 0 - 80 samples, class 1 - 5 samples and class 2 - 125 samples
gini(samples)
```

```python
samples = [0, 0, 125]
gini(samples)
```

```python
def entropy(samples):
    sum_samples = sum(samples)
    entropy = 0
    probabilities = []
    for sample in samples:
        probability = sample / sum_samples
        probabilities.append(probability)
    for probability in probabilities:
        if probability > 0:
            entropy += probability * math.log(probability) 
    return entropy * -1
```

```python
samples = [80, 5, 125]
entropy(samples)
```

```python
samples = [0, 0, 125]
entropy(samples)
```
