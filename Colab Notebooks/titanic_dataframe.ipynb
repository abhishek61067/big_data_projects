```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from google.colab import drive
drive.mount('/content/drive')
```

```python
path = '/content/drive/My Drive/Colab Notebooks/train.csv'
```

```python
titanic_df = pd.read_csv(path)
```

```python
titanic_df.head()
```

**Histogram**


```python
sns.histplot(data = titanic_df,  x = 'Age', bins = 10)
plt.show()

```

**Boxplot chart**

```python
sns.boxplot(data = titanic_df , x = 'Fare')
plt.show()
```

**Scatter diagram**

```python
sns.scatterplot(data = titanic_df, x = 'Age', y = 'Fare', marker =  'o')
```

**Bar plot**

```python
sns.barplot( data = titanic_df,x =  'Age' , y = 'Fare' , color = 'r')
```

**line plot**

```python
sns.lineplot(data = titanic_df, x ="Age", y= 'Fare', color = 'g')
```

**heatmap**


```python
cor = titanic_df.corr()
sns.heatmap(cor, cmap="RdYlBu")
```
